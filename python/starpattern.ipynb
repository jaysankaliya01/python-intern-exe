```python
n = 7
size = 2 * n + 1

for i in range(size):
    for j in range(size):
        # distance from center
        d = max(abs(i - n), abs(j - n))
        
        if d <= n:
            print(n - d, end=" ")
        else:
            print(0, end=" ")
    print()
```

```python
n = 6
size = 2 * n + 3   # gives 15

for i in range(size):
    for j in range(size):

        # border
        if i == 0 or i == size-1 or j == 0 or j == size-1:
            print(0, end=" ")

        # top-left triangle
        elif i <= n and j <= n and j <= n - i:
            print(n - j, end=" ")

        # top-right triangle
        elif i <= n and j > n+1 and j - (n+1) <= i:
            print(j - (n+1), end=" ")

        # bottom-left triangle
        elif i > n+1 and j <= n and i - (n+1) <= j:
            print(i - (n+1), end=" ")

        # bottom-right triangle
        elif i > n+1 and j > n+1 and max(i, j) < size - 1:
            print(min(i, j) - (n+1), end=" ")

        else:
            print(" ", end=" ")
    print()
```

```python
n = 6
size = 2 * n + 3   # 15

for i in range(size):
    for j in range(size):

        # Outer border
        if i == 0 or i == size - 1 or j == 0 or j == size - 1:
            print(0, end=" ")

        # Top-left triangle
        elif i <= n and j <= n + 1:
            val = n - (j - 1)
            if val >= n - i:
                print(val, end=" ")
            else:
                print(" ", end=" ")

        # Top-right triangle
        elif i <= n and j > n + 1:
            val = j - (n + 1)
            if val <= i:
                print(val, end=" ")
            else:
                print(" ", end=" ")

        # Bottom-left triangle
        elif i > n + 1 and j <= n + 1:
            val = i - (n + 1)
            if val <= j - 1:
                print(val, end=" ")
            else:
                print(" ", end=" ")

        # Bottom-right triangle
        elif i > n + 1 and j > n + 1:
            val = j - (n + 1)
            if val <= size - i - 2:
                print(val, end=" ")
            else:
                print(" ", end=" ")

        else:
            print(" ", end=" ")
    print()
```
