```python
n = 5

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
n = 5

for i in range(1, n):
    print(" " * (n - i), end="")
    if i == 1:
        print("*")
    else:
        print("*" + " " * (2*i - 3) + "*")

print("*" * (2*n - 1))
```

```python
n = 5

# Upper half
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2*i - 1))

# Lower half
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "*" * (2*i - 1))
```
