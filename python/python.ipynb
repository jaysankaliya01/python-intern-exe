```python
n = 7

for i in range(n):
    for j in range(n):
        # border condition
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print(0, end=" ")
        else:
            # top-left
            if i <= j and i <= n-1-j:
                print(n-1-i, end=" ")
            # top-right
            elif j < i and j < n-1-i:
                print(n-1-j, end=" ")
            # bottom-left
            elif i > j and i > n-1-j:
                print(i, end=" ")
            # bottom-right
            else:
                print(j, end=" ")
    print()
```

```python
n = 7
mid = n // 2

for i in range(n):
    for j in range(n):
        # border
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print(0, end=" ")

        # top-left
        elif i <= mid and j <= mid:
            print(mid - min(i, j), end=" ")

        # top-right
        elif i <= mid and j > mid:
            print(j - mid, end=" ")

        # bottom-left
        elif i > mid and j <= mid:
            print(i - mid, end=" ")

        # bottom-right
        else:
            print(max(i, j) - mid, end=" ")
    print()
```
