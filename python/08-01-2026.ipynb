```python
#:one: Write a recursive function to reverse a string.
#You must not use slicing ([::-1]) or loops.
def rev(j):
    if j == "":
        return j
    return rev(j[1:]) + j[0]
print(rev("jay sankaliya"))

    
```

```python
#Write a recursive function to calculate the sum of all elements in a list.
list1 = [1, 2, 3, 4, 5]
def sum(list1):
    if len(list1) == 0:
        return 0
    return list1[0] + sum(list1[1:])
print(sum(list1))
```

```python
#Write a recursive function to count how many times a target number appears in a list.
def count_occurrences(lst, target):
    if not lst:
        return 0
    count = 1 if lst[0] == target else 0
    return count + count_occurrences(lst[1:], target)
print(count_occurrences([1, 2, 3, 2, 4, 2, 5], 2))  # Output: 3
```

```python
#Write a recursive function to check whether a string is a palindrome
#Clean the string (ignore spaces + case) inside the function.
def is_palindrome(s):
    s = s.replace(" ", "").lower()
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])
```

```python
#Write a recursive function to find the maximum element in a list.
#You cannot use max() or loops.
def funmax(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        max_of_rest = funmax(lst[1:])
        return lst[0] if lst[0] > max_of_rest else max_of_rest

print(funmax([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))  # Output: 9
```
