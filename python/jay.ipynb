while loops


```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

for loops


```python
fruits = ["limbu", "king", "jay","alon"]
for x in fruits:
  print(x)
  if x == "jay":
    break
  print(x*2)
```
