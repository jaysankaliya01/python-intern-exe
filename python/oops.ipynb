```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def study(self):
        print(self.name," study in india")
s1= Student("ram", 20)
s1.study()
```

```python
class Mobil:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    def call(self):
        print("Calling from", self.brand, self.model)
    def info(self):
        print(f"{self.brand} {self.model} launched in {self.year}")
m1= Mobil("Samsung", "Galaxy S21", 2021)
m1.call()
m1.info()
```

```python
class Bankaccount:
    def __init__(self,balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def deposit(self, amount):
        self.__balance += amount
        print("deposit successful")
acc = Bankaccount(1000)

print(acc.get_balance)
acc.deposit(500)
print(acc.get_balance())
```
